```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
with open("news_story.txt","r") as f:
    news_text = f.read()
    
news_text[:500]
```

```python
news_text
```

```python
doc = nlp(news_text)



for token in doc:
    print(token," | ", token.pos_,"|",spacy.explain(token.pos_))
```

```python
filtered_NOUN = []

for token in doc:
    if token.pos_ in ["NOUN"]:
        filtered_NOUN.append(token)
```

```python
filtered_NOUN
```

```python
filtered_NUM = []

for token in doc:
    if token.pos_ in ["NUM"]:
        filtered_NUM.append(token)
```
